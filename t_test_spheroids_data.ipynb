```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
spheroids_data = pd.read_csv('https://raw.githubusercontent.com/VsevolodMisyurin/csv_library/main/spheroids_20may_2024.csv')

sample_1 = spheroids_data['GPC6_high'].tolist()
sample_2 = spheroids_data['GPC6_low'].tolist()

# Applying two-sample t-test for dependent samples:

from scipy import stats as st
import numpy as np

alpha = 0.05  # Defining significance level

# Applying two-sample t-test for dependent samples
results = st.ttest_rel(sample_1, sample_2,
                       alternative = "greater") # less / greater/ two-sided

# Displaying p-value and t-statistic
print('p-value:', results.pvalue)
print('t-statistic:', results.statistic)

# According to the obtained results, the null hypothesis is rejected.
if results.pvalue < alpha:
    print('Null hypothesis is rejected')
else:
    print('Null hypothesis is not rejected')
```

```python
spheroids_data_melted = spheroids_data.melt(var_name='Condition', value_name='Spheroids')

plt.figure(figsize=(8, 6))
sns.boxplot(x='Condition', y='Spheroids', data=spheroids_data_melted)
plt.title('Boxplot for GPC6_high and GPC6_low')
plt.show()
```
